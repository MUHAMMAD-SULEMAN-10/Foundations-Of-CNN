```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((16, 16), dtype=int)

image[4, 4:12] = 1

image[5:8, 4] = 1


image[8, 4:12] = 1


image[9:12, 11] = 1

image[12, 4:12] = 1
```

```python
image
```

```python
def show_image(img, title):
    plt.imshow(img, cmap="gray", vmin=0, vmax=1)
    plt.title(title)

    h, w = img.shape

    plt.xticks(np.arange(w))
    plt.yticks(np.arange(h))

    plt.gca().set_xticks(np.arange(-0.5, w, 1), minor=True)
    plt.gca().set_yticks(np.arange(-0.5, h, 1), minor=True)


    plt.grid(which="minor", color="gray", linestyle="-", linewidth=0.5)

    plt.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
    plt.show()
```

```python
plt.figure(figsize=(6,6))
show_image(image, "Image")
```

```python
vertical_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

horizontal_kernel = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])
```

```python
def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    output = np.zeros((h - kh + 1, w - kw + 1))

    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            region = image[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)

    return output
```

```python
vertical_features = convolve2d(image, vertical_kernel)
horizontal_features = convolve2d(image, horizontal_kernel)
```

```python
show_image(vertical_features, "veritical")
```

```python

show_image(horizontal_features, "Horizontal")
```
